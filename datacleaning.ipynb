```python
user_map = {
    user: idx
    for idx, user in enumerate(df["user_id"].unique())
}
```

```python
anime_map = {
    anime: idx
    for idx, anime in enumerate(df["anime_id"].unique())
}
```

```python
df["user_index"] = df["user_id"].map(user_map)

df["anime_index"] = df["anime_id"].map(anime_map)
```

```python
df["rating"] = df["rating"] / 10.0
```
